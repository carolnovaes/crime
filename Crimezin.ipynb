```python
### Chamando bases

import pandas as pd
base = pd.read_csv('basecrime.csv', delimiter = ',', encoding = "utf-8")
base
```

```python
base
```

```python
### Tipo de crimes por bairro, por ano e por mês
base["count"] = 1
painel = base.groupby(['ATENDIMENTO_BAIRRO_NOME','ATENDIMENTO_ANO', 'OCORRENCIA_MES', 'NATUREZA1_DESCRICAO']).count()["count"].reset_index(name='count')
painel
```

```python
# convert the 'Date' column to datetime format 
painel['ATENDIMENTO_ANO'] = painel['ATENDIMENTO_ANO'].astype('int') 
painel
```
